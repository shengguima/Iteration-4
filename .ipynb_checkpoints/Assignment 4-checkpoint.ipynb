```python
#Function for getting number of outliers
def count_outliers(df, column, z_threshold=3):
   
    stats = df.select(mean(col(column)).alias("mean"), stddev(col(column)).alias("stddev")).collect()[0]
    mean_value = stats["mean"]
    stddev_value = stats["stddev"]

    outliers_count = df.filter((col(column) < mean_value - z_threshold * stddev_value) |
                               (col(column) > mean_value + z_threshold * stddev_value)).count()
    return outliers_count

#Function for getting number of extremes
def count_extremes(df, column, z_threshold=3.5):
    stats = df.select(mean(col(column)).alias("mean"), stddev(col(column)).alias("stddev")).collect()[0]
    mean_value = stats["mean"]
    stddev_value = stats["stddev"]

    extremes_count = df.filter((col(column) < mean_value - z_threshold * stddev_value) |
                                (col(column) > mean_value + z_threshold * stddev_value)).count()
    return extremes_count

#Function for coercing outliers
def coerce_outliers(df, column, z_threshold=3):
    stats = df.select(mean(col(column)).alias("mean"), stddev(col(column)).alias("stddev")).collect()[0]
    mean_value = stats["mean"]
    stddev_value = stats["stddev"]

    lower_bound = mean_value - z_threshold * stddev_value
    upper_bound = mean_value + z_threshold * stddev_value

    df = df.withColumn(column, 
                       when(col(column) < lower_bound, lower_bound)
                       .when(col(column) > upper_bound, upper_bound)
                       .otherwise(col(column)))
    
    return df

#Function for deleting extremes
def remove_extremes(df, column, z_threshold=3.5):
    stats = df.select(mean(col(column)).alias("mean"), stddev(col(column)).alias("stddev")).collect()[0]
    mean_value = stats["mean"]
    stddev_value = stats["stddev"]

    lower_bound = mean_value - z_threshold * stddev_value
    upper_bound = mean_value + z_threshold * stddev_value

    df = df.filter((col(column) >= lower_bound) & (col(column) <= upper_bound))
    
    return df
```

```python
print("Outlier counts:", count_outliers(df, "Chol"))
print("Extremes counts:", count_extremes(df, "Chol"))
```

```python
df = coerce_outliers(remove_extremes(df,"Age"), "Age")
df = coerce_outliers(remove_extremes(df, "BMI"), "BMI")
df = coerce_outliers(remove_extremes(df, "Chol"), "Chol")
df = coerce_outliers(remove_extremes(df, "TG"), "TG")
df = coerce_outliers(remove_extremes(df, "HDL"), "HDL")
df = coerce_outliers(remove_extremes(df, "LDL"), "LDL")
df = coerce_outliers(remove_extremes(df, "Cr"), "Cr")
df = coerce_outliers(remove_extremes(df, "BUN"), "BUN")
```

```python
age_cla_mapping = {
    'young': 0,
    'middle-aged': 1,
    'elderly': 2
}

BMI_cla_mapping = {
    'underweight': 0,
    'normal': 1,
    'overweight': 2,
    'obese': 3
}

df = df.withColumn("Age_Cla", 
                   when(col("Age_Cla") == 'young', age_cla_mapping['young'])
                   .when(col("Age_Cla") == 'middle-aged', age_cla_mapping['middle-aged'])
                   .when(col("Age_Cla") == 'elderly', age_cla_mapping['elderly'])
                   .otherwise(None))

df = df.withColumn("BMI_Cla", 
                   when(col("BMI_Cla") == 'underweight', BMI_cla_mapping['underweight'])
                   .when(col("BMI_Cla") == 'normal', BMI_cla_mapping['normal'])
                   .when(col("BMI_Cla") == 'overweight', BMI_cla_mapping['overweight'])
                   .when(col("BMI_Cla") == 'obese', BMI_cla_mapping['obese'])
                   .otherwise(None))

df.show()
```

```python
#Discard redundant attributes
df.drop("Age", "BMI")
```

```python
#Normalization
feature_columns = df.toPandas().columns.tolist()

assembler = VectorAssembler(inputCols=feature_columns, outputCol="features")
df_assembled = assembler.transform(df)

scaler = StandardScaler(inputCol="features", outputCol="scaled_features", withMean=True, withStd=True)
scaler_model = scaler.fit(df_assembled)
df_scaled = scaler_model.transform(df_assembled)

def vector_to_array(vector):
    return vector.toArray().tolist()

vector_to_array_udf = udf(vector_to_array, ArrayType(DoubleType()))

df_scaled = df_scaled.withColumn("scaled_features_array", vector_to_array_udf(col("scaled_features")))

for i, col_name in enumerate(feature_columns):
    df_scaled = df_scaled.withColumn(col_name, col("scaled_features_array")[i])
    
df_scaled = df_scaled.select([col_name for col_name in feature_columns])
    
df_final = df_scaled.toPandas()
```

```python
#Visulise the result of normalization
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

df.toPandas().plot(kind='hist', ax=ax[0], bins=30, title='Original Data', alpha=0.7)
df_final.plot(kind='hist', ax=ax[1], bins=30, title='Standardized Data', alpha=0.7)
plt.show()
```

```python
#Find out the best number of features building the model
pca = PCA(k=7, inputCol="features", outputCol="pcaFeatures")
model = pca.fit(df_scaled)

explained_variance = model.explainedVariance.toArray()

cumulative_variance = explained_variance.cumsum()

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='--')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Scree Plot')
plt.grid(True)
plt.show()
```

```python
pc_matrix = model.pc.toArray()
rows = [Row(feature=feature, pc_values=list(pc_matrix[:, idx])) for idx, feature in enumerate(df_scaled.columns)]
pc_df = spark.createDataFrame(rows)
pc_df.show(truncate=False)
```
